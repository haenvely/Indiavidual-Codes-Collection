### NaN 처리하기 
    1. 없는 데이터(결측치)가 있는지 확인한다. 
         - info함수를 통하여 개수를 확인한다.

3. 없는 데이터(NaN)를 특정 값으로 일괄 변경한다.
    - fillna() 함수를 이용해서 특정값으로 결측치를 대체한다.

```python
#컬럼 지정해서  그 기준으로삭제하기. 여러개 선택 가능
df_emp.dropna(subset = ['MANAGER_ID', 'DEPARTMENT_ID']).shape 
```

```python
df_emp['COMMISSION_PCT'].fillna(1)  #복사본이 채워짐.
```

```python
df_emp.info()
```

```python
df_emp['COMMISSION_PCT'].mean()
```

```python
df_emp['COMMISSION_PCT'].fillna(df_emp['COMMISSION_PCT'].min())
```

#### 연습문제
#부서가 80번 부서인데 커미션이 NaN이라면 80번 부서의 평균 커미션을 넣어주고
#80번 부서가 아닌 사원의 NaN인 커미션은 80번 부서의 커미션 최솟값으로 채워준다.

```python
comm1 = df_emp[df_emp['DEPARTMENT_ID'] == 80]['COMMISSION_PCT'].mean()
comm2 = df_emp[df_emp['DEPARTMENT_ID'] != 80]['COMMISSION_PCT'].mean()
print(comm1)
print(comm2)
```

```python
df_emp[(df_emp['DEPARTMENT_ID'] == 80.0) & df_emp['COMMISSION_PCT'].isna()].fillna(df_emp['COMMISSION_PCT'].mean())
```

```python
df_emp[(df_emp['DEPARTMENT_ID'] == 80.0) & df_emp['COMMISSION_PCT'].isnull()]
```

```python
df_emp[(df_emp['DEPARTMENT_ID'] != 80.0) & df_emp['COMMISSION_PCT'].isna()].fillna(df_emp['COMMISSION_PCT'].min())
```

### 특정 키 값을 기준으로 데이터 합치기 
    - groypby() : SQL의 GROUP BY와 동일하다. 특정 컬럼을 기준으로 그룹핑한다. 
    - sum() : 그룹으로 되어있는 데이터를 합친다. 

```python
student_list = [
    {'name':'kang','sex':'female','major':'Computer Science'},
    {'name':'hana','sex':'male','major':'python'},
    {'name':'hong','sex':'male','major':'java'},
    {'name':'esther','sex':'female','major':'Computer Science'},
    {'name':'kim','sex':'female','major':'java'},
    {'name':'nara','sex':'male','major':'python'},
    {'name':'lee','sex':'female','major':'Computer Science'},
    {'name':'haha','sex':'female','major':'java'},
    {'name':'hohee','sex':'male','major':'python'},
    {'name':'leejo','sex':'female','major':'Computer Science'}
]
df = pd.DataFrame(student_list,columns=['name','sex','major'])
df
```

```python
major_group = df.groupby('major') #객체 생성
```

```python
major_group.groups
```

```python
type(major_group)
```

```python
for name, group in major_group:
    print(name, ':', len(group))
    print(group)  #그룹의 정보
    print(type(group))   #그룹이 각 데이터프레임으로 만들어짐
```

```python
sex_group = df.groupby('sex')
```

```python
sex_group.groups
```

```python
for a, b in sex_group:
    print(a, ': ', len(b))
    print(b)
```

```python
#그룹핑을 하는데, 성별만 다른 kang이 두개 있을 때!!

student_list = [
    {'name':'kang','sex':'female','major':'Computer Science'},
    {'name':'hana','sex':'male','major':'python'},
    {'name':'hong','sex':'male','major':'java'},
    {'name':'esther','sex':'female','major':'Computer Science'},
    {'name':'kim','sex':'female','major':'java'},
    {'name':'nara','sex':'male','major':'python'},
    {'name':'lee','sex':'female','major':'Computer Science'},
    {'name':'haha','sex':'female','major':'java'},
    {'name':'hohee','sex':'male','major':'python'},
    {'name':'leejo','sex':'female','major':'Computer Science'},
    {'name': 'kang', 'sex':'male', 'major':'Computer Science'}
]
df = pd.DataFrame(student_list,columns=['name','sex','major'])
df
```

#### apply 함수를 사용하여 사용자 함수 적용하기

```python
date_list = [
    {'yyyy-mm-dd': '2002-10-10'},
    {'yyyy-mm-dd': '2012-11-11'},
    {'yyyy-mm-dd': '2020-05-25'}
]
df = pd.DataFrame(date_list , columns = ['yyyy-mm-dd'])
df
```

```python
#날짜에서 연도만 뽑아서 리턴하는 함수
def get_year(date):
    return date[:4]
get_year('2020-11-11')
```

```python
df['year'] = df['yyyy-mm-dd'].apply(get_year)
```

```python
df
```

#### index를 이용한 group by

   - level은 index의 depth를 의미하며, 가장 왼쪽부터 0부터 증가
