### 1. 두개의 데이터프레임 연결하기

```python
import pandas as pd
```

```python
#연결할 데이터 프레임을 생성한다. 

list1 = [
    {'name':'kang', 'lang':'python'},
    {'name':'kim', 'lang':'java'},
    {'name':'hong', 'lang':'c'},
]

list2 = [
    {'name':'esther', 'lang':'go'},
    {'name':'lee', 'lang':'R'},
    {'name':'park', 'lang':'cobol'},
]
df1 = pd.DataFrame(list1, columns=['name','lang'])
df2 = pd.DataFrame(list2, columns=['name','lang'])
```

### 2. 두 데이터프레임 합치기(병합하기)

```python
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'customer_id': [1, 2, 3],
    'customer_name': ['kang', 'kim', 'hong']
}, columns=['id', 'customer_id', 'customer_name'])
df1
```

```python
df2 = pd.DataFrame({
    'id': [1, 2, 4],
    'order_id': [100, 200, 300],
    'order_date': ['2021-01-21', '2021-02-03', '2020-10-01']
}, columns=['id', 'order_id', 'order_date'])
df2
```

```python
pd.concat([df1, df2], axis = 1)  # 짝이 없는 데이터도 index기준으로 붙여버림
```

```python
pd.merge(df1, df2)  #merge는 짝이 없는 값은 없음
```

```python
df_emp = pd.read_csv('employees.csv')
df_emp
```

```python
df_dept = pd.read_csv('dept.csv')
df_dept
```

#### merge(): 두 데이터프레임을 합치기
- merge(데이터프레임1, 데이터프레임2) : 두 데이터프레임에 동일한 이름을 가진 컬럼을 기준으로 두 데이터프레임을 합침

```python
pd.merge(df_emp, df_dept)
```

- merge(데이터프레임1, 데이터프레임2, on=기준컬럼명) : 기준 컬럼을 명시할 수도 있음

```python
pd.merge(df_emp, df_dept, on = 'DEPARTMENT_ID')
```

#### merge() 를 통해 어떻게 두 데이터프레임을 결함시킬 것인가에 대해 보다 상세한 기능을 제공함
- merge(데이터프레임1, 데이터프레임2, how=결합방법) 
- 결합방법
  1. inner : 내부 조인 - SQL의 INNER JOIN 과 동일
  2. outer : 완전 외부 조인 - SQL의 OUTER JOIN 과 동일
  3. left : 왼쪽 우선 외부 조인 - SQL의 LEFT OUTER JOIN 과 동일
  4. right : 오른쪽 우선 외부 조인 - SQL의 RIGHT OUTER JOIN 과 동일

```python
pd.merge(df_emp, df_dept, on = 'DEPARTMENT_ID', how = 'inner').shape
```

```python
pd.merge(df_emp, df_dept, on = 'DEPARTMENT_ID', how = 'outer').shape
```

```python
pd.merge(df_emp, df_dept, on = 'DEPARTMENT_ID', how = 'left').shape
```

```python
pd.merge(df_emp, df_dept, on = 'DEPARTMENT_ID', how = 'right').shape
```
