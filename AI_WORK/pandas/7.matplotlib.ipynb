#### 여러 그래프 그리기
- 그래프를 보여주는 창을 Figure 라고 한다. 
- 기본은 하나의 Figure에 그래프가 하나 그려진다. 
- 하나의 그래프창에 여러개의 그래프를 그릴 수 있다.
    - 연속해서 plot()함수를 수행해서 여러개의 그래프를 그릴 수 있다.
    - plot()함수에 표시하고자 하는 데이터를 모두 입력해서 여러개의 그래프를 그릴 수 있다.

````python
import numpy as np

x  = np.arange(-4.5, 5, 0.5)

y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2 + 10
````

````python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
````

````python
plt.plot(x, y1, x, y2, x,y3)
plt.show()
````

````python
plt.plot(x, y1)
````

````python
plt.plot(x, y1)
plt.figure(2)  # 하나의 그래프 창 하나하나로 띄어서 만들어줌.
plt.clf()  #지워주는 역할.


plt.plot(x, y2)
plt.figure(1)
plt.clf()
plt.plot(x, y3)
plt.figure()   # 안에 숫자가 순서를 말함.???
plt.show()
````

````python
y4 =  4*x**2 + 5
````

````python
plt.subplot(2,2,1)  # 2 X 2짜리의 첫번째 포지션
plt.plot(x, y1) 

plt.subplot(2,2,2)  # 2 X 2짜리의 2번째 포지션
plt.plot(x, y2) 

plt.subplot(2,2,3)  # 2 X 2짜리의 2번째 포지션
plt.plot(x, y3) 

plt.subplot(2,2,4)  # 2 X 2짜리의 2번째 포지션
plt.plot(x, y4) 
````

````python
plt.plot(x, y1, x, y4)
plt.show()
````

````python
plt.plot(x, y1, x, y4)
plt.xlim(-2, 2)
plt.ylim(5, 35)   #xlim, ylim을 줘서 특정 부분만 크게 상세히 보고싶을 때.
plt.show()
````

###  그래프 꾸미기

#### 출력 형식 지정
   - fmt 옵션을 이용해서 그래프 색깔, 선스타일, 마커등을 지정 할 수 있다.
   - fmt = '[color][line_style][marker]'
   - 교재 298 참고

````python
x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
````

````python
plt.plot(x, y1, x, y2, x, y3, x, y4)  # 별다른 값을 주지 않으면 기본값으로 지정됨.
plt.show()
````

````python
plt.plot(x, y1,'k', x, y2, x, y3, x, y4, 'c')   # y1과 y4만 지정색. 나머진 기본색.
plt.show()
````

````python
plt.plot(x, y1,':', x, y2, x, y3, x, y4, '-.')   # y1과 y4만 선 모양 바뀜.
plt.show()
````

````python
#y1 색이랑 마커 같이 지정가능: o + m
plt.plot(x, y1,'om', x, y2,'^', x, y3,'s', x, y4, 'c') 
plt.show()
````

#### 라벨, 제목, 격자, 범례, 문자열

````python
plt.plot(x, y1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plot table title')
plt.grid()  #격자찍기  #plt.grid(True)

plt.show()
````

````python
plt.plot(x, y1,'om--', x, y2,'^', x, y3,'s', x, y4, 'c') 

plt.legend(['y1', 'y2', 'y3', 'y4'])
plt.show()
````

````python
plt.plot(x, y1,'om', x, y2,'^', x, y3,'s', x, y4, 'c') 

plt.legend(['y1', 'y2', 'y3', 'y4'], loc = 'lower right')  #범주의 위치
plt.show()
````

### 산점도
- 두 연속형 변수 간의 관계를 확인하고 두 연속형 변수 사이에 상관관계가 있는지 확인할 때 사용한다.
- 예를 들면, 날씨와 아이스크림 판매량의 관계, 매출과 직원수의 관계, 시험점수와 공부시간의 관계등을 시각화 할때 유용하다. 



https://www.jmp.com/support/help/ko/14-2/dj-visualize-6.shtml

````python
height = [165, 177, 160, 180, 185, 155, 172] #키
weight = [62, 67, 55, 74, 90, 43, 64] #몸무게
````

````python
plt.scatter(height, weight)
plt.xlabel('키 데이터(cm)')
plt.ylabel('몸무게 데이터(kg)')
plt.title('키와 몸무게의 산점도 Scatter Plot')
plt.grid(True)
plt.show()
````

````python
plt.scatter(height, weight, s = 500, c = 'r')  #사이즈와 컬러 지정
plt.show()
````

````python
size = 100 * np.arange(1, 8)  # 점마다 다른 크기 지정
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y']

plt.scatter(height, weight, s = size, c = colors)
plt.show()
````
