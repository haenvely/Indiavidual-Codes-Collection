## 학습목표
 1. 숫자, 범주형 데이터를 이해한다. 
 2. 숫자 데이터를 범주형 데이터로 바꿀 수 있다.
 3. 범주형 데이터를 전처리 할 수 있다.

```python
import pandas as pd
```

```python
# https://www.kaggle.com/hesh97/titanicdataset-traincsv/data
train_data = pd.read_csv('./train.csv')
train_data.head()
```

#### info함수로 각 변수의 데이터 타입 확인
 - 타입 변경은 astype함수를 사용

```python
train_data.info()
```

#### 숫자형(Numerical Type) 데이터
 - 연속성을 띄는 숫자로 이루어진 데이터이다.
   - 예) Age, Fare 등

#### 범주형(Categorical Type) 데이터
 - 연속적이지 않은 값(대부분의 경우 숫자를 제외한 나머지 값)을 갖는 데이터를 의미한다.
   - 예) Name, Sex, Ticket, Cabin, Embarked 
 - 어떤 경우, 숫자형 타입이라 할지라도 개념적으로 범주형으로 처리해야할 경우가 있다.
   - 예) Pclass

```python
train_data['Pclass'] = train_data['Pclass'].astype(str)
```

```python
train_data.info()
```

```python
pd.get_dummies(train_data).info()
```

```python
#범주형을 몽땅 0/1로 전환해줌. Pclass, Cabin 등등.

# 기계학습시 범주형 데이터를 학습시킬 때 0/1로 구분해주기 위해서.
pd.get_dummies(train_data)  
```
