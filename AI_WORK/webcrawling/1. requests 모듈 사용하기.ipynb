#### **requests 모듈**
1. http의 요청(rquest), 응답(response)을 위한 모듈이다. 

```python
import requests
```

#### get 요청하기
1. http get방식으로 요청한다.
2. 쿼리문자열을 이용하여 데이터를 전달한다.

```python
url = 'https://www.yonhapnewstv.co.kr/news/MYH20201229002200038'
requests.get(url)
```

```python
#response 객체로 받았으니 할당해서 보자.
resp = requests.get(url)
resp.status_code #상태코드가 응답에 넘어옴. 200번대면 정상
#없는 아이디 넣으면 404에러..400번대면 클라이언트 잘못. 500번대면 서버에서 동작하다가 에러.
```

```python
resp.text #페이지에서 직접 소스보기 했을 때 내용과 동일
```

#### post 요청하기
1. http post방식으로 요청한다.
2. http  body를 이용하여 데이터 전달한다.

```python
url = 'https://www.kangcom.com/member/member_check.asp'
#url에 붙이는 방식이 아니기 때문에 요청에 넘겨줄 데이터를 url이 아닌 바디에 넘겨야함.
mydata = {
    'id':'carami',
    'pwd':'tttt'
}
resp = requests.post(url, data = mydata)
resp.text
```

#### HTTP header 데이터 이용하기
1. header 데이터 구성하기
2. header 데이터 전달하기

```python
#정보를 제대로 가져오지 못한다면..?
#헤더 정보를 넘겨줄 수 도 있다. 헤더에서 더 많은 정보를 얻어내고 싶다면?
#ajax 이용 - 데이터만 보내는 형식의 홈페이지. 즉, api에 있는 데이터를 요청해야함.

#권한을 가지고 있는 클라이언트에게만...
# f12 - network - name - haders - user agent
#user-agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
}
url = 'https://www.yonhapnewstv.co.kr/news/MYH20201229002200038'
resp = requests.get(url, headers = headers) 
resp.text
```

#### HTTP response 처리하기
1. response 객체의 이해
2. status_code 확인하기
3. text 속성 확인하기

```python
url = 'https://www.yonhapnewstv.co.kr/news/MYH20201229002200038'
resp = requests.get(url)
if resp.status_code == 200: #요청을 정상으로 처리했다면
    print(resp.text)
else:
    print('error')
```
