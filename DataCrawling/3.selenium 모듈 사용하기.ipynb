#### HTTP 상태 코드
 - 1xx (정보): 요청을 받았으며 프로세스를 계속한다
 - 2xx (성공): 요청을 성공적으로 받았으며 인식했고 수용하였다
 - 3xx (리다이렉션): 요청 완료를 위해 추가 작업 조치가 필요하다
 - 4xx (클라이언트 오류): 요청의 문법이 잘못되었거나 요청을 처리할 수 없다
 - 5xx (서버 오류): 서버가 명백히 유효한 요청에 대해 충족을 실패했다

#### selenium을 활용하여 특정 element의 로딩 대기
 - WebDriverWait 객체를 이용하여 해당 element가 로딩 되는 것을 대기
 - 실제로 해당 기능을 활용하여 거의 모든 사이트의 크롤링이 가능
 - WebDriverWait(driver, 시간(초)).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'CSS_RULE')))

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(chrome_driver)

url = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=105&oid=081&aid=0003018031'
# chrome driver 로 해당 페이지가 물리적으로 open
driver.get(url)

# 해당 element가 로딩 될때까지 대기
#myElem = WebDriverWait(driver, 10)
WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CSS_SELECTOR, '.u_cbox_count')))

src = driver.page_source

soup = BeautifulSoup(src)
comment_area = soup.select_one('.u_cbox_count')

# chrome driver 사용 후, close 함수로 종료
driver.close()


print(comment_area.get_text())
```

### 네이버 웹툰 - Best 댓글 크롤링하기
예시: 수요웹툰 중 고수
고수 1화의 베스트 댓글을 크롤링해보자.

##### 유의사항
댓글은 웹툰 페이지 URL로 접근 불가능하다.
CSS - Network에서 comment의 request url을 찾아서 접근해야한다.


```python
from bs4 import BeautifulSoup as bs
import time

chrome_driver = 'C:/Users/ohi02/PycharmProjects/AI_WORK/chromedriver'
driver = webdriver.Chrome(chrome_driver)
cmurl = 'https://comic.naver.com/comment/comment.nhn?titleId=662774&no=1'
driver.get(cmurl)

time.sleep(3)
source = driver.page_source
comment = bs(source)

#print(comment)
best_comment = comment.find('span', class_ = 'u_cbox_contents')
best_id = comment.find('span', class_='u_cbox_nick')
best_like = comment.find('em', class_ = 'u_cbox_cnt_recomm')
print(best_comment.get_text())
print(best_id.get_text())
print(best_like.get_text())

driver.close()
```

#### 베스트 댓글을 리스트로 전부 가져와보자!

```python
from bs4 import BeautifulSoup as bs
import time

chrome_driver = 'C:/Users/ohi02/PycharmProjects/AI_WORK/chromedriver'
driver = webdriver.Chrome(chrome_driver)
cmurl = 'https://comic.naver.com/comment/comment.nhn?titleId=662774&no=1'
driver.get(cmurl)

time.sleep(3)
source = driver.page_source
comment = bs(source)

#print(comment)
best_comment = comment.find_all('span', class_ = 'u_cbox_contents')
best_id = comment.find_all('span', class_='u_cbox_nick')
best_like = comment.find_all('em', class_ = 'u_cbox_cnt_recomm')
```

#### 튜플로 구성된 리스트로 만들어보자!

[(댓글내용, 작성자, 좋아요수), (댓글내용2, 작성자2, 좋아요수2) , , , , ]



```python
best = list(zip(best_comment, best_id, best_like))
print(best[0][0].get_text())
print(best[0])
```

###### 그중 가장 높은 좋아요수를 받은 댓글은 뭘까???

```python
best_like_new = [int(i.get_text()) for i in best_like]
print(best_like_new)
print(int(best_like_new[0]))
```

```python
max(best_like_new)
```

```python
for i,b in enumerate(best):
    if int(b[-1].get_text()) == max(best_like_new):
        print('가장 Like를 많이 받은 댓글입니다')
        print('아이디명: ', b[1].get_text())
        print('댓글 내용: ', b[0].get_text())
        print('Like 수: ', b[-1].get_text())
```
